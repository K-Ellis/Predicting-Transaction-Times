```python
import time
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os  # Used to create folders
from shutil import copyfile  # Used to copy parameters file to directory
from sklearn.metrics import mean_squared_error
import math

def get_parameters(parameters):
    d = {}
    with open(parameters, "r") as f:
        for line in f:
            line = line.replace(":", "")
            line = line.replace(",", "")
            line = line.split()
            key = line.pop(0)
            if len(line) > 1:
                d[key] = line
            else:
                d[key] = line[0]
    return d

parameters = "../../../Data/parameters.txt"  # Parameters file
sample_parameters = "../Sample Parameter File/parameters.txt"

d = get_parameters(parameters)
sample_d = get_parameters(parameters)

for key in sample_d.keys():
    if key not in d.keys():
        print("\"%s\" not in parameters.txt - please update parameters file with this key" % key)
        print("Default key and value => \"%s: %s\"" % (key, sample_d[key]))
        exit()
for key in d.keys():
    if key not in sample_d.keys():
        print("\"%s\" not in sample parameters" % key)
```

```python
df = pd.read_csv(d["file_location"] + d["plotting_input_file"] + ".csv", encoding='latin-1',
                              low_memory=False)
df["TimeTaken"] /= 3600
```

```python
lastx= 0
ave_timetaken = []
std_timetaken = []
for x in range(1,298):
    x*2
    ave_timetaken.append(df["TimeTaken"][(df["Cases_created_within_past_8_hours"]>=lastx) & (df["Cases_created_within_past_8_hours"]<x)].mean())
    std_timetaken.append(df["TimeTaken"][(df["Cases_created_within_past_8_hours"]>=lastx) & (df["Cases_created_within_past_8_hours"]<x)].std())
    lastx=x
ave_timetaken = np.array(ave_timetaken)
std_timetaken = np.array(std_timetaken)
# plt.plot([x for x in range(297)],ave_timetaken)
plt.errorbar([x for x in range(297)], ave_timetaken,yerr=std_timetaken, fmt="r", ecolor="b", elinewidth=0.5, capsize=1000, errorevery=10)
# plt.plot([x for x in range(297)],ave_timetaken+std_timetaken)
# plt.plot([x for x in range(297)],ave_timetaken-std_timetaken)
plt.show()
```

```python
lastx= 0
ave_timetaken = []
std_timetaken = []
for x in range(1,int(298/10)):
    x*2
    ave_timetaken.append(df["TimeTaken"][(df["Cases_created_within_past_8_hours"]>=lastx) & (df["Cases_created_within_past_8_hours"]<x)].mean())
    std_timetaken.append(df["TimeTaken"][(df["Cases_created_within_past_8_hours"]>=lastx) & (df["Cases_created_within_past_8_hours"]<x)].std())
    lastx=x
ave_timetaken = np.array(ave_timetaken)
std_timetaken = np.array(std_timetaken)
# plt.plot([x for x in range(28)],ave_timetaken)
plt.errorbar([x for x in range(28)], ave_timetaken,yerr=std_timetaken, fmt="r", ecolor="b", elinewidth=0.5, capsize=1000, errorevery=1)
plt.show()
```
