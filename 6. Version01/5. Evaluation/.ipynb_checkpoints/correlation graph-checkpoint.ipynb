```python
import networkx as nx
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../../Data/B4.csv", encoding="latin1")
# keepers = ["Cases_created_within_past_8_hours",
#            "Concurrent_open_cases",
#            "Created_on_Weekend",
#            "Seconds_left_Day",
#            "Seconds_left_Month",
#            "Seconds_left_Qtr",
#            "TimeTaken"]
#            #, "Created_On", "ResolvedDate"]
# for col in df:
#     if col not in keepers:
#         del df[col]
df.head()
```

```python
cor_matr = df.corr()
cor_matr.fillna(0, inplace=True)
cor_matr
```

```python
G = nx.Graph()
npcor_matr = np.array(cor_matr)
for i in range(len(npcor_matr)):
    for j in range(len(npcor_matr)):
        if i > j:
            if abs(npcor_matr[i][j]) > 0.3:
#             print(npcor_matr[i][j])
#             print(cor_matr.columns[i], cor_matr.columns[j])
                G.add_edge(cor_matr.columns[i], cor_matr.columns[j],weight = abs(npcor_matr[i][j]))
nx.write_gexf(G, "../../../Data/cor_matrix_recommended/cor_matrix_less_edges.gexf")
```

```python
del df["TimeTaken"]
G = nx.Graph()
npcor_matr = np.array(cor_matr)
for i in range(len(npcor_matr)):
    for j in range(len(npcor_matr)):
        if i > j:
#             print(npcor_matr[i][j])
#             print(cor_matr.columns[i], cor_matr.columns[j])
            G.add_edge(cor_matr.columns[i], cor_matr.columns[j],weight = abs(npcor_matr[i][j]))
nx.write_gexf(G, "../../../Data/cor_matrix_recommended/cor_matrix_no_TimeTaken.gexf")
```
