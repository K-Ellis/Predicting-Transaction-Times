### plot RMSE to day, week, month, qtr, year

```python
df.head()
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np
```

```python
def get_daily_error(df, alg):
    df['weekday'] = df['Created_On'].apply(lambda x: x.weekday())
    r2_scores = []
    rmse_scores = []
    mae_scores = []
    for i in range(7):
        actual = df.loc[df["weekday"] == i, "TimeTaken_hours"]
        predicted = df.loc[df["weekday"] == i, "TimeTaken_hours_%s"%alg]
        r2_scores.append(r2_score(actual, predicted))
        rmse_scores.append(np.sqrt(mean_squared_error(actual, predicted)))
        mae_scores.append(mean_absolute_error(actual, predicted))
    daily_scores = [r2_scores, rmse_scores, mae_scores]
    return daily_scores

def plot_daily_errors(daily_score, error_name, alg,y_label):
    days_of_week = ["Su", "Mo", "Tu", "We", "Th", "Fr", "Sa"]
    days_of_week_num = [i for i in range(len(days_of_week))]
    plt.plot(days_of_week_num, daily_score)
    plt.xticks(days_of_week_num, days_of_week)
    plt.title("%s - %s to Day of the Week"% (alg, error_name))
    plt.ylabel(y_label)
    plt.xlabel("Day of the Week")
    plt.show()

def plot_errors(df, alg):
    df["Created_On"] = pd.to_datetime(df["Created_On"])
    df["ResolvedDate"] = pd.to_datetime(df["ResolvedDate"])
    df["TimeTaken_hours"] = df["TimeTaken"]/60/60
    df["TimeTaken_hours_%s"%alg] = df["TimeTaken_%s"%alg]/60/60
    daily_scores = get_daily_error(df, alg) 
    error_names = ["R2", "RMSE", "MAE"]
    y_labels = ["R2 score", "RMSE (Hours)", "MAE (Hours)"]
    for daily_score, error_name, y_label in zip(daily_scores, error_names, y_labels):
        plot_daily_errors(daily_score, error_name, alg, y_label)
    
plot_errors(df, "LinearRegression")
```
