# Frequencies of values in each column

# Queue

```python
print(df["Queue"].value_counts())

prob = df["Queue"].value_counts()
threshold = 100
mask = prob > threshold
tail_prob = prob.loc[~mask].sum()
prob = prob.loc[mask]
prob['other'] = tail_prob

plt.figure(figsize=(20,7))

prob.plot(kind='bar')
plt.xticks(rotation=90, fontsize = 20)
plt.grid()
plt.show()
```

# StatusReason

```python
print(df["StatusReason"].value_counts())

prob = df["StatusReason"].value_counts(normalize = True)

prob.plot(kind='bar')
plt.xticks(rotation=90, fontsize = 20)
plt.grid()
plt.show()
```

```python
cols = list(df.columns[7:])
cols.remove("CaseRevenue")
cols.remove("RelatedCases")
cols.remove("CreditAmount")
cols.remove("DebitAmount")
cols.remove("OrderAmount")
cols.remove("InvoiceAmount")
cols.remove("Deleted")
print(cols)

for col in cols:
    print("---------------------------\nColumn = ", col,"\n---------------------------\n")
    print(df[col].value_counts())
    
    if len(df[col].unique())<=11:
        prob = df[col].value_counts(normalize = True)
        
        plt.figure(figsize=(20,5))
        prob.plot(kind='bar')
        plt.xticks(rotation=90, fontsize = 20)
        plt.grid()
        plt.show()
        
    elif len(df[col].unique())<=200:
        prob = df[col].value_counts(normalize = True)
        top_10_prob = df[col].value_counts(normalize = True)
        top_10_prob[:10] = True
        top_10_prob[10:] = False
        
        mask = top_10_prob
        mask = mask.astype(np.bool)
        tail_prob = prob.loc[~mask].sum()
        prob = prob.loc[mask]
        prob['other'] = tail_prob

        plt.figure(figsize=(20,5))
        prob.plot(kind='bar')
        plt.xticks(rotation=90, fontsize = 20)
        plt.grid()
        plt.show()

    else:
        print("\nDidn't plot", col, "because of too many unique values\n")
```
