```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# pd.set_option("display.max_columns",70)
dfcleaned = pd.read_csv("../../../Data/vw_Incident_cleaned.csv", encoding='latin-1')
dforiginal = pd.read_csv("../../../Data/vw_Incident.csv", encoding='latin-1', low_memory=False)
dforiginal.AmountinUSD.fillna(0, inplace=True)
dfs = [dfcleaned.AmountinUSD, dforiginal.AmountinUSD]
df_names = ["cleaned.AmountinUSD", "original.AmountinUSD"]
for df, name in zip(dfs, df_names):
    print("--------------------\n", name, "\n--------------------")
    print(df.describe(), "\n")
```

The original AmountinUSD ranges from -2,373,858 to 55,586,710. The scaled AmountinUSD ranges from 0 to 1. Therefore values which were 0 in the original df get scaled up to >0 in the cleaned df and the smallest value in the original df (-2,373,858) then gets scaled to 0 in the cleaned df.

```python
for df, name in zip(dfs, df_names):
    print("--------------------\n", name, "\n--------------------")
    plt.plot(df)
    plt.show()
```

```python
for df, name in zip(dfs, df_names):
    print("--------------------\n", name, "\n--------------------")
    df.hist(bins="auto")
    plt.show()
```
