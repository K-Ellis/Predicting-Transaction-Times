```python
import pandas as pd
import numpy as np
import datetime
import time
from calendar import monthrange
from pprint import pprint
```

```python
df = pd.read_csv("../../../Data/vw_Incident.csv",encoding='latin-1',low_memory=False)
```

```python
from sklearn.utils import resample
df = resample(df, n_samples=100, random_state=1234)
```

```python
dfcopy = df.copy()
cols = ["TimeTaken", "rolling_mean", "rolling_median", "rolling_std", "Created_On", "ResolvedDate"]
for col in df:
    if col not in cols:
        del dfcopy[col]
        
dfcopy["Created_On"] = pd.to_datetime(df["Created_On"])
dfcopy["ResolvedDate"] = pd.to_datetime(df["ResolvedDate"])
dfcopy["TimeTaken"] = (dfcopy["ResolvedDate"] - dfcopy["Created_On"]).astype('timedelta64[s]')

dfcopy.sort_values(by="Created_On", inplace=True)

dfcopy["rolling_mean"] = dfcopy["TimeTaken"].rolling(window=10).mean()
dfcopy["rolling_median"] = dfcopy["TimeTaken"].rolling(window=10).median()
# dfcopy["rolling_max"] = dfcopy["TimeTaken"].rolling(window=10).max()
# dfcopy["rolling_min"] = dfcopy["TimeTaken"].rolling(window=10).min()
dfcopy["rolling_std"] = dfcopy["TimeTaken"].rolling(window=10).std()
dfcopy["rolling_sum"] = dfcopy["TimeTaken"].rolling(window=10).sum()


dfcopy.dropna(subset=["rolling_mean"], inplace=True)
dfcopy[cols]
```
