```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dfincident = pd.read_csv("../../../Data/vw_Incident_cleaned_with_TicketNumber.csv",
                      encoding='latin-1',
                     low_memory=False)
dfholdactivity = pd.read_csv("../../../Data/vw_HoldActivity_cleaned.csv",
                      encoding='latin-1',
                     low_memory=False)
dfaudithistory = pd.read_csv("../../../Data/vw_AuditHistory_cleaned.csv",
                      encoding='latin-1',
                     low_memory=False)
dfpackagetiageetry = pd.read_csv("../../../Data/vw_PackageTriageEntry_cleaned.csv",
                      encoding='latin-1',
                     low_memory=False)
```

```python
unique_tickets = dfholdactivity["TicketNumber"].unique()
ABT = pd.DataFrame(dfholdactivity["TicketNumber"].unique(), columns=["TicketNumber"])
ABT["HoldDuration"] = 0
for ticket in unique_tickets:
    duration = dfholdactivity.loc[dfholdactivity["TicketNumber"] == ticket, 'HoldDuration'].sum()
    ABT.loc[ABT["TicketNumber"] == ticket, "HoldDuration"] = duration
dfincident = dfincident.merge(ABT,how='left', left_on='TicketNumber', right_on='TicketNumber')
dfincident["HoldDuration"].fillna(0, inplace=True)
# dfincident.to_csv("../../../Data/vw_Incident_cleaned_HoldDuration.csv", index=False)
```

for each column, if the join ticket == 1, create a row in dfbinary with a 1

```python
def show_existance_of_variable(dftransactions, cols, dfincident):
    unique_tickets = dftransactions["TicketNumber"].unique()
    ABT = pd.DataFrame(unique_tickets, columns=["TicketNumber"])
    for col in cols:
        ABT[col] = 0
        for ticket in unique_tickets:
            exists = dftransactions.loc[dftransactions["TicketNumber"] == ticket, col].sum()
            if exists > 0:
                ABT.loc[ABT["TicketNumber"] == ticket, col] = 1
            else:
                ABT.loc[ABT["TicketNumber"] == ticket, col] = 0
#         ABT[col].fillna(0, inplace=True)
    ABT = ABT.merge(dfincident, how='right', left_on='TicketNumber', right_on='TicketNumber')
#     dfincident = dfincident.merge(ABT,how='left', left_on='TicketNumber', right_on='TicketNumber')

#     ABT.to_csv("../../../Data/ABT.csv", index=False)
    return ABT
```

```python
dftransactions = dfholdactivity
cols = ["Reason_3PP - Identify Payment", "Reason_3PP - Other", "AssignedToGroup_Internal", "AssignedToGroup_Submitter (Contact)"]
ABT = show_existance_of_variable(dftransactions,cols, dfincident)
```

```python
holdactivitycolumns = dfholdactivity.columns.tolist()
not_included = ["TicketNumber","ActivityId", "RegardingObjectId", "HoldDuration", "TimeZoneRuleVersionNumber", "HoldTypeName_Customer"]
for item in not_included:
    holdactivitycolumns.remove(item)

dftransactions = dfholdactivity
ABT1 = show_existance_of_variable(dftransactions,holdactivitycolumns, dfincident)
ABT1
```

```python
dfaudithistorycolumns = dfaudithistory.columns.tolist()
not_included = ["TicketNumber","AuditHistoryId", "AuditId", "EntityId", "CaseId", "UTCConversionTimeZoneCode", "Created_On", "Modified_On", 
               "Action_Update"]
for item in not_included:
    dfaudithistory.remove(item)

dftransactions = dfaudithistory

ABT2 = show_existance_of_variable(dftransactions,dfaudithistorycolumns, ABT1)
ABT2
```

```python
packagetiageetrycolumns = dfpackagetiageetry.columns.tolist()
not_included = ["TicketNumber","PackageTriageEntryId", "PackageTriageFormId", "Created_On", "Modified_On"]
for item in not_included:
    packagetiageetrycolumns.remove(item)

dftransactions = dfpackagetiageetry
ABT3 = show_existance_of_variable(dftransactions,packagetiageetrycolumns, ABT2)
dfholdactivity[["AssignedToGroup_Internal"]]
ABT3
```
