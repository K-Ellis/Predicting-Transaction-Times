```python
extrinsic =["Concurrent_open_cases",
"Cases_created_within_past_8_hours",
"Cases_resolved_within_past_8_hours",
"Seconds_left_Day",
"Seconds_left_Month",
"Seconds_left_Qtr",
"Created_on_Weekend",
"Rolling_Mean",
"Rolling_Median",
"Rolling_Std"]
```

```python
for df in dfs:
    df["Intrin"] = False
    for val in df["Columns"].values:
        if val not in extrinsic:
            # then intrinsic = True
            df["Intrin"].loc[df["Columns"] == val] = True
```

```python
total_in = []
total_ex = []
mean_in = []
mean_ex = []
std_in = []
std_ex = []
col = "Percentage_Importance"
i=0
for df in dfs:
    total_in.append(df[col][df["Intrin"]].sum())
    total_ex.append(df[col][df["Intrin"]==False].sum())
    mean_in.append(df[col][df["Intrin"]].mean())
    mean_ex.append(df[col][df["Intrin"]==False].mean())
    std_in.append(df[col][df["Intrin"]].std())
    std_ex.append(df[col][df["Intrin"]==False].std())
    i+=1
    if i>1:
        col = "Importances"
```

```python
print(total_in)
print(total_ex)
print(mean_in)
print(mean_ex)
print(std_in)
print(std_ex)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from matplotlib.ticker import FuncFormatter
```

```python
total_ex
```

```python
fig, ax = plt.subplots()
sns.barplot([x for x in range(len(total_ex))], total_ex, palette="Blues")
plt.ylabel("Total % Importance for Generated Variables")
plt.xlabel("Algorithms")
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 
xticks=["LR", "EN", "GBR", "RFR"]
plt.xticks([x for x in range(len(xticks))], xticks)
for i, v in enumerate(total_ex):
    print(v, i)
    ax.text(i-0.2, v+0.01, "%.4f%%"%v, color='Black', fontweight='bold')#, fontsize=15)
plt.title("Total % Importance for Generated Variables for each Algorithm")
plt.savefig("../../../Data/Importances/B4Total_pct_for_each_alg.png")
plt.show()
```

```python
fig, ax = plt.subplots()
sns.barplot([x for x in range(len(mean_ex))], mean_ex, palette="Blues")

plt.ylabel("Average % Importance for Generated Variables")
plt.xlabel("Algorithms")
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 
xticks=["LR", "EN", "GBR", "RFR"]
plt.xticks([x for x in range(len(xticks))], xticks)
for i, (v, std) in enumerate(zip(mean_ex, std_ex)):
    print(v, i)
    ax.text(i-0.2, v+0.01, "%.4f%%"%v, color='Black', fontweight='bold')#, fontsize=15)
    
    print(std)
    plt.plot([i-0.2, i-0.2], [v-std, v+std])
    
    
plt.title("Total % Importance for Generated Variables for each Algorithm")
plt.show()
```
